```python
# A dog class

class Dog:
    def __init__(self, name, weight, breed, mood):
        self.name = name
        self.weight = weight
        self.breed = breed

    def show(self):
        print(f'Name : {self.name}  Weight : {self.weight}kg    Breed : {self.breed}')

    def bark(self):
        if self.mood == "Nervous":
            print("Woof")
    
billy = Dog("Billy", 20.0, "German Shepherd", "Nervous")
print(type(billy))
print(billy.breed)
```

```python
# Public and Private members


# Public members

class Dog:
    def __init__(self, name, weight, breed):
        self.name = name
        self.weight = weight
        self.breed = breed

billy = Dog("Billy", 20.0, "German Shepherd")
print(billy.breed)
print(billy.weight)
print(billy.name)
```

```python
# Private members

class Dog:
    def __init__(self, name, weight, breed):
        self.name = name
        self.__weight = weight
        self.__breed = breed

    def show(self):
        print(f'Name : {self.name}  Weight : {self.__weight}kg    Breed : {self.__breed}')


billy = Dog("Billy", 20.0, "German Shepherd")
billy.show()
print(billy.__breed)
print(billy.__weight)
print(billy.name)
```

```python
# Composite Classes

# Example

class Department:
    def __init__(self):
        self.department_name = ""
        self.students = 0

    def setDepartmentName(self, dep_name):
        self.department_name = dep_name
    
    def getDepartmentName(self):
        return self.department_name

    def setStudents(self, stu_num):
        self.students = stu_num
    
    def getStudents(self):
        return self.students

class University:
    def __init__(self, number_departments):
        self.departments = [Department() for _ in range(number_departments)]


    def setDepartmentName(self):
        print("Enter the department names ")
        for i in range(len(self.departments)):
            self.departments[i].department_name = input()

    def addStudents(self,dep,stu_num):
        for i in range(len(self.departments)):
            if self.departments[i].department_name == dep:
                self.departments[i].students += stu_num
    
    def setStudents(self, dep, stu_num):
        for i in range(len(self.departments)):
            if self.departments[i].department_name == dep:
                self.departments[i].students = stu_num

    def showInfo(self):
        for department in self.departments:
            print(f'Department : {department.department_name}   Students : {department.students}')
    

uowm = University(2)
uowm.setDepartmentName()
uowm.addStudents()
uowm.showInfo()
```

```python
# Static class members

# Example 
class User:
    users = 0

    def __init__(self):
        User.users += 1

u1 = User()
u2 = User()

print(User.users)
print(u1.users)
print(u2.users)
```

```python
# The 'del' keyword

# Example 1

l = [1,2,3]
del l[0]
print(l)

d = {1:1, 2:2}
del d[1]
print(d)
```

```python
# Example 2

class User:
    users = 0

    def __init__(self):
        User.users += 1

    def __del__(self):
        User.users -= 1

u1 = User()
u2 = User()
print(u1.users)
del u1
print(u2.users)
```

```python
# Object Equality

class MyClass:
    def __init__(self, num):
        self.member = num

o1 = MyClass(1)
o2 = MyClass(2)
o3 = o2

print(id(o1), id(o2), id(o3))
print(o1 is o2)
print(o2 is o3)
print()

print([1,2] is [1,2])
print([1,2] == [1,2])
print()

o4 = MyClass(1)
print(o1 == o4)     # Error. We have to overload the == operator to work with objects
```

```python
# isinstance() Method

print(isinstance([1,2], list))      # is [1,2] a list     
print(isinstance({"a":2}, list))      # is {"a":2} a dictionary
print(isinstance(o4, MyClass))      # is 'o4' an object of MyClass
print(isinstance([1,2], (list,int)))        # is [1,2] a list or an integer
```
